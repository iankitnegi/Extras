```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("bar_chart_data.csv").sort_values("Cars Listings")
df
```

```python
plt.barh(df["Brand"], df["Cars Listings"], color="midnightblue")
plt.title("Car Listing by Brands", fontsize=14, fontweight="bold")
plt.ylabel("Brands")
plt.xlabel("No. of Cars Listing")
plt.show()
```

```python
df.plot(x="Brand", y="Cars Listings",
        kind='barh')
plt.title("Car Listing by Brands", fontsize=14, fontweight="bold")
plt.legend('', frameon=False)
plt.ylabel("Brands")
plt.xlabel("No. of Cars Listing")
plt.show()
```

```python
list = ["#1D2F6F", "#8390FA", "#FAC748", "#F9E9EC", "#F88DAD"]
custom_palette = sns.color_palette(list)
sns.set_palette(custom_palette)
```

```python
df = pd.read_csv("bar_chart_homework_data.csv")
df = df.iloc[0:3, 0:3]
```

```python
plt.bar(df["Cities"], df["Frequency"])
plt.title("Charts by Cities X Frequency", fontsize=14, fontweight="bold")
plt.ylabel("Frequency")
plt.xlabel("Cities")
plt.show()
```

```python
df_fuel_engine_type = pd.read_csv("pie_chart_data.csv")
df_fuel_engine_type
```

```python
plt.pie(df_fuel_engine_type["Number of Cars"], 
        labels=df_fuel_engine_type["Engine Fuel Type"], 
        autopct="%0.2f%%",
        textprops={"size": "x-large", "fontweight": "bold", "rotation":30, "color": "w"})
plt.legend()
plt.show()
```

```python
df = pd.read_csv("stacked_area_chart_data.csv")
df
```

```python
c = ["#011638", "#7e2987", "#8390FA"]
plt.figure(figsize=(18,4))
sns.set_style("white")
plt.stackplot(df["Year"],  
              df["Diesel"], 
              df["Petrol"],
              df["Gas"],
              colors=c,
              edgecolor="none")
plt.xticks(df["Year"], rotation=45)
plt.legend(labels= ["Disel", "Petrol", "Gas"], loc="upper left")
plt.title("Popularity of Engine Fuel Types (1982-2016)", fontsize=14, weight="bold")
plt.ylabel("Numbers of Cars")
sns.despine()
plt.show()
```
