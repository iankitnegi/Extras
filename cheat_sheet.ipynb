```python
# Each VARIABLES is saved in its own COLUMNS, Each OBSERVATION is saved in its own ROWS.
import pandas as pd
```

```python
# Creating Data Frames using dic -> Col

pd.DataFrame(
    {
        "Name": ["John", "Gopal", "Jack", "Vishu"],
        "Age": [10, 56, 19, 22],
        "Occupation": ["Business Analyst", "Data Analyst", "Sales Manager", "Consultant"]
    }
)
```

```python
#Creating DataFrames using list -> Rows

pd.DataFrame(
    [
        ["John", 10, "Business Analyst"],
        ["Gopal", 56, "Data Analyst"],
        ["Jack", 19, "Sales Manager"],
        ["Vishnu", 22, "Consulatnt"]
    ], index=[0,1,2,3], columns=["Name", "Age", "Occupation"]
)
```

```python
df = pd.DataFrame(
    {
        "A":[1,2],
        "B":[3,4],
        "C":[5,6],
        "D":[7,8]
    }
)
df
```

```python
dataframe = pd.melt(df, var_name="Str", value_name="Int")
```

```python
df = pd.pivot(dataframe, columns="Str", values="Int")
df
```

```python
df1 = pd.DataFrame(
    {
        "A": [1,2], "B": [3,4], "C":[5,6]
    }
)
df1
```

```python
df2 = pd.DataFrame(
    {
        "A": [7,8], "B": [9,10], "C":[11,12]
    }
)
df2
```

```python
df3 = pd.DataFrame(
    {
        "D": [7,8], "E": [9,10], "F":[11,12]
    }
)
df3
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df3], axis=1)
```
